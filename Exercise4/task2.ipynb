```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plotting the Bifurcation Diagram

```python
alphas = np.linspace(0, 1, 100)

plt.plot(alphas,  np.sqrt(alphas), color="green", label="stable")
plt.plot(alphas, -np.sqrt(alphas), color="red", label="unstable")
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.title(r"Fixed Points of the dynamical system $\dot{x} = \alpha - x^2$")
plt.grid(True)
plt.ylabel(r"$x^*$")
plt.xlabel(r"$\alpha$")
plt.legend()
plt.show()
```

```python
alphas = np.linspace(3, 4, 100)

plt.plot(alphas,  np.sqrt((alphas - 3) / 2), color="green", label="stable")
plt.plot(alphas, -np.sqrt((alphas - 3) / 2), color="red", label="unstable")
plt.xlim(2, 4)
plt.ylim(-1, 1)
plt.title(r"Fixed Points of the dynamical system $\dot{x} = (\alpha - 3) - 2x^2$")
plt.grid(True)
plt.ylabel(r"$x^*$")
plt.xlabel(r"$\alpha$")
plt.legend()
plt.show()
```
