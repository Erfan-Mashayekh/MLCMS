## Task 1: Vector fields, orbits, and visualization

```python
# define notebook parameters
w = 5
Y, X = np.mgrid[-w:w:100j, -w:w:100j]

# example matrix
alphas = [-2, -0.9, -0.2, 0.2, 0.9, 2]
matrices = []
for alpha in alphas:
    A = np.array([[alpha, -2*alpha**3],
                  [2*alpha, -3*alpha]])
    matrices.append(A)
```

```python
def plot(matrix):
    # example for Euler's method to construct and plot a trajectory over the stream plot
    y0 = np.array([-1, 1.5])
    time = np.linspace(0, 10, 1000)
    yt, time = solve_euler(lambda y: matrix@y, y0, time)
    
    # example linear vector field A*x
    ax0 = plot_phase_portrait_linear(matrix, X, Y)
    # then plot the trajectory over it
    ax0.plot(yt[:, 0], yt[:, 1], c='red', label='Example trajectory')
    ax0.set_xlim(-5, 5)
    ax0.set_ylim(-5, 5)
    # prettify
    ax0.legend()
    ax0.set_aspect(1)
    plt.show()
```

```python
for matrix in matrices:
    print(f'Jacobian matrix: \n {matrix}')
    plot(matrix)
```
